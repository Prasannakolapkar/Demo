```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/insurance.csv")
```

```python
df['sex'] = df['sex'].replace("male",1).replace("female",0)
df['smoker'] = df['smoker'].replace("yes",1).replace("no",0)
df['region'] = df['region'].replace("southwest",1).replace("southeast",0).replace("northwest",2).replace("northeast",3)
```

```python
plt.figure(figsize=(10,10))
plt.hist(df['charges'], bins =30)
plt.show()
```

```python
x = df['bmi']
y = df['charges']

plt.scatter(x,y)
plt.xlabel("bmi")
plt.ylabel("charges")
plt.title("bmi vs charges")
plt.show()
```

```python
x = df['region']
y = df['charges']

sns.boxplot(x=x,y=y)
plt.xlabel("region")
plt.ylabel("charges")
plt.title("charges vs region")
plt.show()
```
